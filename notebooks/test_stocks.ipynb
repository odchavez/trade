```python
from trade import stocks as s
from datetime import datetime, timedelta
```

```python
stock_obj = s.Stocks("GOOG")
today = datetime.today().date()
ten_days_ago = today - timedelta(days=20)

future_expiration_date = str(today + timedelta(days=3))

days_interval = stock_obj.days_until(date_str=future_expiration_date)

stock_obj.get_historical_returns(interval=days_interval, start=ten_days_ago, end=today)
```
